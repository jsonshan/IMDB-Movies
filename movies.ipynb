Classifying Movies Based on Ratings: [Exceptional, Good, Average, Not Good]

```python
def rating(rating):
  if rating>=9.0:
    return "Excellent"
  elif rating >=8.2:
    return "Good"
  elif rating >= 7.75:
    return "Average"
  else:
    return "Not Good"
```

```python
movies['rating_cat']=movies['IMDB_Rating'].apply(rating)
```

Unique Values From Genre Column

```python
Genres=[]
for value in movies['Genre']:
  Genres.append(value.split(','))
```

```python
Genres
```

```python
AllGenreSplit=[]
for item in Genres:
  for item1 in item:
    AllGenreSplit.append(item1)
```

```python
AllGenreSplit
```

```python
uni_list=[]
for item in AllGenreSplit:
  if item not in uni_list:
    uni_list.append(item)
```
